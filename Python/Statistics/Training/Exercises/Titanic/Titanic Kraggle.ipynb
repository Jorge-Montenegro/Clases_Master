```python
#Esta función genera un Tabla de Frecuencias para cualquier DataFrame y nombre de columna que no sea un string/object
def tabla_de_frecuencias(dframe,columna):
    frecuencia=pd.crosstab(index=dframe[columna],columns=columna)
    frecuencia['Frec_Abs']=pd.value_counts(dframe[columna])
    #Esto ordena por el indice - frecuencia.sort_index()
    frecuencia['Frec_Rel']=frecuencia['Frec_Abs']/len(dframe[columna])
    frecuencia['Frec_Por']=frecuencia['Frec_Rel']*100
    frecuencia['Frec_Abs_Acum']=frecuencia['Frec_Abs'].cumsum()
    frecuencia['Frec_Rel_Acum']=frecuencia['Frec_Abs_Acum']/len(dframe[columna])
    return frecuencia

#Programa que devuelve todas las frecuencias
tabla_frecuencias=[]
for i in titanic.columns:
    #OJO, para que no sume valores que no numéricos hay que poner una condición como esta
    if titanic[i].dtype!=np.object:
        tabla_frecuencias.append(tabla_de_frecuencias(titanic,i))

#Ahora vamos a listar cada una de las tablas de Frecuencias        
```
