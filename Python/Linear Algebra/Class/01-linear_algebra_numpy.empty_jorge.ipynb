## A Linear Regression example with numpy

Now, we are ready to implement our own linear regression example. 

In linear regression, our hypothesis function $h_\theta$ is:

$$h_\theta(x) = \theta_0 + \theta_1x$$

And, as we are doing regression, our cost function is: 

$$J(\theta_0,\theta_1) = \frac{1}{m}\sum_{i=1}^m(\hat{y}_i-y_i)^2 = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i)^2 $$

### Generate dummy data

```python
#Nuestra función será y=50 + 1,3*x + un valor aleatorio de la muestra
theta_real_0=50
theta_real_1=1.3

xs=np.random.normal(30.0,10.0, 250)
ys=theta_real_0 + theta_real_1 * xs + np.random.normal(0,10.0, 250)

plt.scatter(xs,ys)
```

```python
#Valores iniciales de theta_0, theta_1
theta_real_0=50
theta_real_1=1.3
#Valores de X
xs=np.random.normal(30.0,10.0, 250)
#Función real con los valores de X más un valor aleatorio
ys=theta_real_0 + theta_real_1 * xs + np.random.normal(0,10.0, 250)

#Tupla con los valores de theta
theta_prima_0=0
theta_prima_1=0

#Factor para el gradiente
alpha=0.0001

#Función de coste
def coste (theta_0,theta_1,ys,xs):
    y_prima=theta_0 + theta_1 * xs
    return np.sum((y_prima-ys)**2)/len(xs)

#Función gradiente que devuelve una Tupla
def gradiente (theta_0,theta_1,ys,xs):
    y_prima=theta_0 + theta_1 * xs
    return np.sum(y_prima-ys)/len(xs),np.sum((y_prima-ys)*xs)/len(xs)

#Inicio de programa
coste_inicial=coste (theta_prima[0],theta_prima[1],ys,xs)

while True:
    #Nuevos valores de theta_prima con el gradiente
    theta_prima_0 = theta_prima_0 - alpha * gradiente(theta_prima_0,theta_prima_1,ys,xs)[0]
    theta_prima_1 = theta_prima_1 - alpha * gradiente(theta_prima_0,theta_prima_1,ys,xs)[1]
    
    #Cálculo del coste de la función
    coste_final=coste(theta_prima_0,theta_prima_1,ys,xs)
    
    #Condición de salida. Que el coste inicial - coste final sea menor a un umbral que he usado el mismo para el alpha
    if abs(coste_inicial - coste_final) < alpha:
        break
    
    #Ahora el coste final es el coste inicial para volver a iterar
    coste_inicial=coste_final

print('Theta_prima_0: %g y Theta_prima_1: %g' %(theta_prima_0,theta_prima_1))    
#Función ajustada
ys_prima=theta_prima_0 + theta_prima_1 * xs
#Pintamos la nube de puntos con la función ajustada a dichos puntos
plt.scatter(xs,ys)    
plt.plot(xs,ys_prima,color='blue')
```

That is the spread that we will try to aproximate with our line.

### Write the cost function

### Gradient descent

Remember, we have to descend in the direction of the steepest gradient. For that, we need to now what direction the gradient points!

### Partial differentials of the cost function

$$\frac{\partial}{\partial\theta_0} = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i)$$

$$\frac{\partial}{\partial\theta_1} = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i) \cdot x_i$$

### Implementing gradient descent

Now we are ready to implement the actual gradient descent. Remember, the algorithm is:

- Initialize variables

- Compute cost function

- Compute gradients

- Update variables: gradient times learning rate (alpha)

- Repeat until convergence: cost at iteration n-1 $\sim$ cost at iteration n
