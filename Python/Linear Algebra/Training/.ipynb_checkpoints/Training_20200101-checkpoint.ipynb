```python
#Ejercicio de los pollos y conejos
```

$35 = x + 2y$

$94 = x + 4y$

$$X = A^{-1} \cdot B$$

```python
matriz_a=np.array(([1,1],[2,4]))
matriz_b=np.array([35,94])
matriz_a_inv=la.inv(matriz_a)
matriz_x=matriz_a_inv.dot(matriz_b)
print("Hay %d pollos y %d conejos"%(matriz_x[0],matriz_x[1]))
```

```python
a = np.array([[1, 1], [2, 4]])
b = np.array([35, 94])
c = np.linalg.solve(a,b)
a
```

```python
np.append(a,[[35,94]],axis=0).reshape(2,3)
```
