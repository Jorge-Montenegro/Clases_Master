## Matplotlib y Numpy

```python
#Importar para plotear
import matplotlib.pyplot as plt
#Importar numpy
import numpy as np
```

```python
#Hay que poner esto para que los resultados de Matplotlib se vean en línea como resultado de células
%matplotlib inline
#Esto es el estilo popular en R
plt.style.use('ggplot')
#si quieres ver el resto de estilos se hace de la siquiente manera
print(plt.style.available)
```

```python
plotX=[1,3,7]
plotY=[7,3,1]
#Dibujo de línea
plt.plot(plotX,plotY)
```

```python
plotX=[1,3,7]
plotY=[7,3,1]
#Dibujo scatter
plt.scatter(plotX,plotY)
```

```python
#Función linspace
#Devuelve por defecto 50 elementos repartidos entre un punto inicial y punto final
#Si ponemos num=X decimos cuantos elementos queremos
#Devuelve un tipo especial numpy.ndarray
np.linspace(2,7, num=100)
```

```python
x=np.linspace(2,8,num=100)
y=np.cos(x)
plt.plot(x,y)
plt.plot(x,np.sin(x))
```

### Arrays con numpy

```python
array=[1,2,3,4]
narray=np.array(array)
#Devuelve una tupla con el número de elementos
narray.shape
#4 Columnas
```

```python
array=[[1,2,3,4],[1,2,3,4]]
narray=np.array(array)
narray.shape
#2 filas y 4 columnas
```

```python
#Funciona igual que linspace pero no toca el final. El step será lo que le diga el usuario
y=np.arange(0,1,0.1)
y
```

```python
#También se puede crear un array de ceros
np.zeros(10)
```

```python
#Y se puede crear una tabla de ceros. Se le pasa un tupla por parámetro
np.zeros((2,5))
```

```python
#Matriz identidad
np.eye(8,8)
```

```python
#Parámetro de entrada una tupla
x=np.ones((2,7))
```

```python
x=x*3
print(x)
```
